```python
# Plotting the values - counts of Numerical features

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
x=0

for i in numeric_column:
    sns.histplot(data=data,x=i,kde=True)
    plt.show()
    
    

```

```python
# to understand the correlation between the numeric features. we cannot do this for categorical features
# we can have a threshold of say 0.5 and wataver feature combinations have more than 0.5, we can drop any one of the feature. 
#This is to avoid the multi colinearity problem, we will try to give more unque features to the models  
data_num = data[numeric_column]
sns.heatmap(data_num.corr(), annot=True)
```

```python
# Plotting the values - counts of Categorical features - we use count plot not Histogram..Histogram is only for numeric

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
x=0

for i in categorical__column:
    sns.countplot(data=data,x=i)
    plt.show()
    
    

```

```python
# any ML model you cannot provide categorical column. so I need to encode the feature and send it to Model

cut_map={'Premium':4, 'Very Good':3, 'Ideal':5, 'Good':2, 'Fair':1}
```

```python

data_cat["clarity"].unique()
```

```python
clarity_map={'VS2':4, 'SI2':2 , 'VS1':5, 'SI1':3, 'IF':8, 'VVS2':6, 'VVS1':7, 'I1':1}
color_map={'F':5, 'J':1, 'G':4, 'E':6, 'D':7, 'H':3, 'I':2}
```

```python
data['cut']=data['cut'].map(cut_map)
data['clarity'] = data['clarity'].map(clarity_map)
data['color'] = data['color'].map(color_map)
```
